## Part I: Web Scraping

This section is about data collection. The <a href="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M">URL</a> provided by the instructors contains a comprehensive list of Canadian postal codes.
We will use the BeautifulSoup library to collect the content of this table, and then generate a computable dataset from it.

The import cell contains:
* requests to cope with HTTP
* BeautifulSoup to extract HTML markups from HTTP responses
* pandas for data processing tasks

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = requests.get(url)
content = page.text
soup = BeautifulSoup(content, 'html.parser')
```

```python
data = soup.table.tbody.find_all('tr')
rows = []
for row in data:
    rows.append(row.text)

rows
```

Rows list already contains the whole dataset.
From now we can:
* get rid of the table headers in rows[0]
* leverage the '\n' characters to structure the dataset

```python
rows.pop(0)
index = []
for r in rows:
    row = []
    row = r.split('\n')
    row.pop(0)
    row.pop(-1)
    for e in row:
        e = e.strip()
    index.append(row)
index
```

1. We aggregate data in the DataFrame table
1. Then sort table rows where both Borough and Neighborhood are 'Not assigned'

```python
ls = index.copy()

table = pd.DataFrame(ls)
table.columns = ['PostalCode','Borough','Neighborhood']
```

```python
bor = table[table['Borough'] == 'Not assigned']
to_del = bor[bor['Neighborhood'] == 'Not assigned']  

len(to_del)
```

We get back to the data in a list format:
1. We duplicate the list ls to keep iteration consistency
1. We iterate backwards in the list duplicate to drop the rows listed in to_del

```python
lst = ls.copy()

for i in range(-1,-78,-1):
    j = to_del.index[i]
    lst.pop(j)
lst
```

Finally, data is stored in a new DataFrame and exported to csv:

```python
conso = pd.DataFrame(lst, columns = ['PostalCode','Borough','Neighborhood'])
conso.to_csv('./toronto_pc_conso.csv')
```
